Как мы видим, когда человек покупает премиум, то он чаще посещает приложение. Пробная подписка тоже, хоть и немного, но стимулирует пользователей чаще заходить на платформу

```python
def func_nan (visit_days, user_type, temperary_days):
    if visit_days == 'nan' and user_type == 'free':
        return users.query('visit_days != "nan" and user_type == "free"')\
                .temperary_days.median()
    elif visit_days == 'nan' and user_type == 'probe':
        return users.query('visit_days != "nan" and user_type == "probe"')\
                .temperary_days.median()
    elif visit_days == 'nan' and user_type == 'payer':
        return users.query('visit_days != "nan" and user_type == "payer"')\
                .temperary_days.median()
    else:
        return temperary_days
```

```python
users['days'] = users.apply(lambda x: func_nan(x.visit_days, x.user_type, x.temperary_days), axis=1)
```

Проверим равенство дисперсий для данных коэффициента привлекаельности для женщин и мужчин.
F-тест используется для проверки равенства двух дисперсий совокупности.
Нулевая и альтернативная гипотезы для теста следующие:

H 0 : σ 1  = σ 2 (дисперсии равны)

H 1 : σ 1 ≠ σ 2 (дисперсии не равны)

```python
def f_test(x, y):
 x = np.array(x)
 y = np.array(y)
 f = np.var(x, ddof=1)/np.var(y, ddof=1) #calculate F test statistic 
 dfn = x.size-1 
 dfd = y.size-1 
 p = 1-scipy.stats.f.cdf(f, dfn, dfd) #find p-value of F test statistic 
 return p
```

```python
x = users.query('gender == "female"').attraction_coeff
y = x = users.query('gender == "male"').attraction_coeff
print('p = ',f_test(x, y))
if f_test(x, y) < 0.05:
    print('дисперсии не равны')
else:
    print('дисперсии равны')
```

Распредлеления attraction_coeff для мужчин и женщин не является нормальным (при этом гомогенность дтсперсий выполняется), поэтому для оценки наличия разницы между значениями конверсии просмотра в лайк для мужчин и женщин воспользуемся непараметрическим статистическим тестом -  U-критерий Манна — Уитни. <br>
Нулевая гипотеза которого: Не имеется различий между двумя независимыми выборками (мужская аудитория и женская аудитория) по уровню коэффициента привлекательности.

```python
U1, p = mannwhitneyu(users.query('gender == "male"').attraction_coeff,
    users.query('gender == "female"').attraction_coeff)
print(f'U1 = {U1}, p = {round(p, 4)}')
```

p < 0.05, следовательно различия между мужчинами и женщинами явлются статистически значимыми. Тем самым мы определили то, что мужчины чаще женщин ставят лайки при просмотре профилей женщин. Этот факт довольно значим для бизнеса, ведь он говорит о том, что женщины чаще получают отклики на свои фотки анкеты нежели мужчины. ## Проведение А/А и A/B тестирований

### Определение метрик:

Для оценки влияния изменения цены премиум подписки будем использовать следующие метрики:

1) Конверсия пользователя с бесплатной подпиской в обладателя платного премиум-аккаунта 

2) Конверсия пользователя из владельца пробной подписки в обладателя платной подписки премиум аккаунта

3) ARPU (Average revenue per user) - для оценки средней выручки одного пользователя 

4) ARPPU (Average revenue per paying user) - для оценки средней выручки одного платящего пользователя 

### А/A тестирования

**I) Конверсия пользователя с бесплатной подпиской в обладателя платного премиум-аккаунта 
и конверсия пользователя из владельца пробной подписки в обладателя платной подписки премиум аккаунта**

```python
tabAA= pd.crosstab(users[users['group'] != 'B' ].user_type, users[users['group'] != 'B' ].group)
tabAA
```

```python
mosaic(tabAA.stack(), gap=0.02);
```

```python
g, p, dof, expctd = chi2_contingency(tabAA)
print('р уровень значимости равен:', p, ', то есть р > 0.05')
```

Таким образом, нельзя отклонить 0 гипотезу об отсутствии влияния контрольных выборок на конверсии. То есть для обоих групп A1 и A2 разница между конверсиями статистически незначимая

Проверим данную бизнес-метрику не один раз, а с помощью синтетических симуляций (300 штук)

```python
simulations = 300 # число симуляций
n_s = 50 # число элементов, берущихся из выборки для проведения единичного t теста
res = []
lower_treshhold = 0

d0 = users.user_type
d1 = users.group

# Запуск симуляций A/A теста
for i in tqdm(range(simulations)):
    s1 = d0.sample(n_s, replace=False).values
    s2 = d1.sample(n_s, replace=False).values
    
    tabAB= pd.crosstab(s1, s2)
    g, p, dof, expctd = chi2_contingency(tabAB)
    res.append(p) # сохраняем pvalue
    if p < 0.05:
        lower_treshhold += 1

plt.hist(res, bins=50)
plt.style.use('ggplot')
plt.xlabel('pvalues')
plt.ylabel('frequency')
plt.title("Histogram of ttest A/A simulations ")
plt.show()
```

```python
# Проверяем, что количество ложноположительных случаев не превышает альфа = 0.05
sum(np.array(res) < 0.05) / simulations
```

Распределение получилось очень близким к равномерному. Это говорит о том, что А/A тестирование не выявило проблем с системой сплитования

**II) ARPU (Average revenue per user) - для оценки средней выручки одного пользователя**


Проверим данную бизнес-метрику не один раз, а с помощью синтетических симуляций (2000 штук)

```python
simulations = 2000 # число симуляций
n_s = 800 # число элементов, берущихся из выборки для проведения единичного t теста
res = []

d0 = users.query('group == "A1"').total_revenue
d1 = users.query('group == "A2"').total_revenue

# Запуск симуляций A/A теста
for i in tqdm(range(simulations)):
    s1 = d0.sample(n_s, replace=False).values
    s2 = d1.sample(n_s, replace=False).values
    res.append(stats.ttest_ind(s1, s2, equal_var=False)[1]) # сохраняем pvalue

plt.hist(res, bins=50)
plt.style.use('ggplot')
plt.xlabel('pvalues')
plt.ylabel('frequency')
plt.title("Histogram of ttest A/A simulations ")
plt.show()
```

```python
# Проверяем, что количество ложноположительных случаев не превышает альфа = 0.05
sum(np.array(res) < 0.05) / simulations
```

Распределение получилось очень близким к равномерному. Это говорит о том, что А/A тестирование не выявило проблем с системой сплитования

```python
print('ARPU A1 =', round(users.query('group == "A1"').total_revenue.mean(), 3))
```

```python
print('ARPU A2 =', round(users.query('group == "A2"').total_revenue.mean(), 3))
```

**III) ARPPU (Average revenue per paying user) - для оценки средней выручки одного платящих пользователей**

Проверим данную бизнес-метрику не один раз, а с помощью синтетических симуляций (2000 штук)

```python
simulations = 2000 # число симуляций
n_s = 50 # число элементов, берущихся из выборки для проведения единичного t теста
res = []
lower_treshhold = 0

d0 = users.query('group == "A1" and total_revenue > 0').total_revenue
d1 = users.query('group == "A2" and total_revenue > 0').total_revenue

# Запуск симуляций A/A теста
for i in tqdm(range(simulations)):
    s1 = d0.sample(n_s, replace=False).values
    s2 = d1.sample(n_s, replace=False).values
    res.append(stats.ttest_ind(s1, s2, equal_var=False)[1]) # сохраняем pvalue
    if stats.ttest_ind(s1, s2, equal_var=False)[1] < 0.05:
        lower_treshhold += 1

plt.hist(res, bins=50)
plt.style.use('ggplot')
plt.xlabel('pvalues')
plt.ylabel('frequency')
plt.title("Histogram of ttest A/A simulations ")
plt.show()
```

```python
# Проверяем, что количество ложноположительных случаев не превышает альфа = 0.05
lower_treshhold / simulations  # доля p значений менее 0.05 
```

Распределение получилось близким к равномерному. Это говорит о том, что А/A тестирование не выявило проблем с системой сплитования

Проверим равенство дисперсий для A и B.
F-тест используется для проверки равенства двух дисперсий совокупности.
Нулевая и альтернативная гипотезы для теста следующие:

Объявим функцию, которая позволит проверять гипотезы с помощью бутстрэпа

```python
def get_bootstrap(
    data_column_1,  # числовые значения первой выборки
    data_column_2,  # числовые значения второй выборки
    boot_it=1000,  # количество бутстрэп-подвыборок
    statistic=np.mean,  # интересующая нас статистика
    bootstrap_conf_level=0.95  # уровень значимости
):
    boot_len = max([len(data_column_1), len(data_column_2)])
    boot_data = []
    for i in tqdm(range(boot_it)):  # извлекаем подвыборки
        samples_1 = data_column_1.sample(
            boot_len, 
            replace=True  # параметр возвращения
        ).values
        
        samples_2 = data_column_2.sample(
            boot_len, 
            replace=True
        ).values
        
        boot_data.append(statistic(samples_1 - samples_2))  # mean() - применяем статистику
        
    pd_boot_data = pd.DataFrame(boot_data)
        
    left_quant = (1 - bootstrap_conf_level)/2
    right_quant = 1 - (1 - bootstrap_conf_level) / 2
    ci = pd_boot_data.quantile([left_quant, right_quant])
        
    p_1 = norm.cdf(
        x=0, 
        loc=np.mean(boot_data), 
        scale=np.std(boot_data)
    )
    p_2 = norm.cdf(
        x = 0, 
        loc = -np.mean(boot_data), 
        scale = np.std(boot_data)
    )
    p_value = min(p_1, p_2) * 2
        
    # Визуализация
    plt.hist(pd_boot_data[0], bins=50)
    plt.style.use('ggplot')
    plt.vlines(ci, ymin=0, ymax=50, linestyle='--')
    plt.xlabel('boot_data')
    plt.ylabel('frequency')
    plt.title("Histogram of boot_data")
    plt.show()
       
    return {"boot_data": boot_data, 
            "ci": ci, 
            "p_value": p_value}
```

```python
sns.set(rc={'figure.figsize': (20, 7)})
x = users.query('total_revenue < 100000 and group == "A"').total_revenue
y = users.query('total_revenue < 100000 and group == "B"').total_revenue
booted_data = get_bootstrap(y, x, boot_it=2000)
```

```python
booted_data["ci"] # местоположение хвостов границ
```

```python
print('p =', round(booted_data["p_value"], 3), '> 0.05' )
```

```python
a = 1 - booted_data["p_value"] # доля значений за пределами границ 
print(a)
```

Мы получили результат, говорящий о том, что данные тестовой и контрольной выборок существенно не различаются. С вероятностью менее 0.20 значение разности средних значений данных по выручке не содержит 0 в своем доверительном интервале. То есть средние значения распределений не различаются значимо. Принимаем 0 гипотезу

**II) ARPU (Average revenue per user) - средняя нормированная выручка на одного пользователя:**

Проверим на нормальность распределения трат и на гомогенность диспрсий трат для А и B.

```python
users1.query('group == "A"').total_revenue.count()
```

```python
stat, p = shapiro(users1.query('group == "A"').total_revenue.sample(n=1000, random_state=1))
print('stat = %.3f, p = %.3f' % (stat, round(p,4)))
if p > 0.05:
    print("Данные трат для А распределены по нормальному распределению")
else:
    print("Данные трат для A распределены не по нормальному распределению")
```

```python
users1.query('group == "B"').total_revenue.count()
```

```python
stat, p = shapiro(users1.query('group == "B"').total_revenue.sample(n=1000, random_state=1))
print('stat = %.3f, p = %.3f' % (stat, round(p,4)))
if p > 0.05:
    print("Данные трат для B распределены по нормальному распределению")
else:
    print("Данные трат для B распределены не по нормальному распределению")
```

Проверим равенство дисперсий для A и B.
F-тест используется для проверки равенства двух дисперсий совокупности.
Нулевая и альтернативная гипотезы для теста следующие:

H 0 : σ 1  = σ 2 (дисперсии равны)

H 1 : σ 1 ≠ σ 2 (дисперсии не равны)

```python
x = users1.query('group == "A"').total_revenue.sample(n=1000, random_state=3)
y = users1.query('group == "B"').total_revenue.sample(n=1000, random_state=3)
print('p = ',f_test(x, y))
if f_test(x, y) < 0.05:
    print('дисперсии не равны')
else:
    print('дисперсии равны')
```

Распределение total_revenue не является нормальным, а их дисперсии являются гомогенными, поэтому для данной задачи пооценки статистической значимости различий можно использовать непараметрический метод Манна-Уитни. Но в бизнесовой сфере более предпочтителен bootstrap для проверки статистических гипотез.

Гипотезы:<br /> 
H0 - изменение цены подписки в двух новых платежных системах не влияет на среднюю выручку на одного пользователя <br />
H1 - влияние имеется.

Проверили медианы и поняли, что средний чек пользователей равен 0. То есть менее 50% пользователей юзают премиум версию.

```python
users1.groupby('group_old')\
    .agg({'total_revenue': 'mean'})\
    .reset_index().rename(columns={'total_revenue': 'ARPU','group_old':'group'})\
    .set_index('group')
```

```python
sns.set(rc={'figure.figsize': (20, 7)})
x = users1.query('total_revenue < 100000 and group == "A"').total_revenue
y = users1.query('total_revenue < 100000 and group == "B"').total_revenue
booted_data = get_bootstrap(y, x, boot_it=2000)
```

Мы получили результат, говорящий о том, что данные тестовой и контрольной выборок существенно не различаются. С вероятностью менее 0.38
значение разности средних значений данных по выручке не содержит 0 в своем доверительном интервале. То есть средние значения распределений не различаются значимо. Принимаем 0 гипотезу

Нулевая и альтернативная гипотезы для теста следующие:
